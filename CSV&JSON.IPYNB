Aggregate Function = Reduce a set of values into a single summary value 
used to summarize and analyze data 
Often used with the groupby() function

```python
#Whole DataFrame
print(df.mean(numeric_only=True))
print(df.min(numeric_only=True))
print(df.max(numeric_only=True))
print(df.sum(numeric_only=True))
print(df.count())
```

```python
#Single column
print(df["Height"].mean(numeric_only=True))
print(df["Height"].min(numeric_only=True))
print(df["Height"].max(numeric_only=True))
print(df["Height"].sum(numeric_only=True))
print(df["Height"].count())

```

```python
#Group by
group=df.groupby("Type1")
print(group["Height"].mean())
print(group["Height"].sum())
print(group["Height"].min())
print(group["Height"].max())
print("\n")
print(group["Height"].count())
```

Data cleaning = the process of fixing/removing:
incomplete,incorrect, or irrevelevant data.
~75% of work done with pandas is data cleaning


```python
df=df.drop(columns=["Legendary","No"])
print(df)
```

```python
#Handle missing data
df=df.dropna(subset=["Type2"])
df=df.fillna({"Type2":"None"})
#Fix incomplete values
df["Type1"]=df["Type1"].replace({"Grass":"GHAS",
                                 "Water":"Pani",
                                 "Fire":"Aago"})
#Standarize text
df["Type1"]=df["Type1"].str.capitalize()
#Fix data types
df["Legendary"]=df["Legendary"].astype(bool)
df=df.drop_duplicates()

print(df.to_string())
```
