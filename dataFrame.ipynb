DataFrame= A tabular data structure with rows and columns.(2 Dimensional)
Similar to an Excel spreadsheet
loc=location by lable

```python
import pandas as pd

```

```python
data={"Name": ["Alice", "Bob", "Charlie"],
      "Age": [25, 30, 35]
      }
df=pd.DataFrame(data,index=["Emp1", "Emp2", "Emp3"])
print(df)
print("\n")
print(df.loc["Emp2"])
print("\n")
print(df.iloc[0])
```

Add a new column and row

```python
df["job"]=["cook","N/A","Teacher"]
print(df)
new_row=pd.DataFrame([{"Name":"Sandy","Age":21,"Job":"Engineer"},
                      {"Name":"Mandy","Age":22,"Job":"Manager"}]
                     ,index=["Emp4","Emp5"])
df=pd.concat([df,new_row])
print("\n")
print(df)
```
