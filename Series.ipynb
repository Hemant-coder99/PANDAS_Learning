```python
import pandas as pd
```

series= A pandas 1-Dimensional labeled array that can hold any data type.
Think of it like a single column in a spreadsheet(1-Dimensional)

```python
data=[100,102,204]
series=pd.Series(data)
print(series)

Apartment=pd.Series(data,index=["AP1","AP2","AP3"])
print(Apartment)

```

Access data using propertiies 

```python
data=["a",102,204]
series=pd.Series(data,index=["a","b","c"])
print(series.loc["a"])
print("\n")
series.loc["c"]="changed value"
print(series)


#interger index
print(series.iloc[2])
```

```python
values=[100,103,340,109,234,764,14,234,654,23]
series=pd.Series(values)
print(series[series>200])
```

dictionary use

```python
calories={"day1":1200,"day2":1500,"day3":1000}
series=pd.Series(calories)
print(series)

series.loc["day2"]+=200
print(series["day2"])
series["day3"]-=100
print(series["day3"])

print(series[series>1300])
```
